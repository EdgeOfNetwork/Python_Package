```python
import matplotlib.pyplot as plt

x = np.arange(0, 6, 0.1) #0에서 6까지 0.1 간격으로 생성
y = np.sin(x)

plt.plot(x, y)
plt.show()
```

```python
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1, label="sin")
plt.plot(x, y2, linestyle="--", label="cos") #cos 함수는 점선으로 그려라
plt.xlabel("x") #name of x axis
plt.ylabel("y")
plt.title('sin & cos') #title
plt.legend()
plt.show()
#https://kongdols-room.tistory.com/87 레전드 사용방법
```
